```python
import pandas as pd
import requests
from plyer import notification
from datetime import datetime

```

```python
url = "https://restcountries.com/v3.1/all"
response = requests.get(url)
```

```python
if response.status_code == 200:
    data_json = response.json()
else:
    print("Erro ao acessar API")
```

```python
if response.status_code == 200:
    notification.notify(
    title = 'Importação API',
    message = f"Import com Sucesso!",
    app_name = 'VSCODE ALERT!',
    timeout = 10)
else:
    notification.notify(
    title = 'Importação API',
    message = f"Erro na importação dos dados",
    app_name = 'VSCODE ALERT!',
    timeout = 10)
```

```python
nome         =[pais['name']['common'] for pais in data_json]
nome_oficial =[pais['name']['official'] for pais in data_json]
region       =[pais['region'] for pais in data_json]
population   =[pais['population'] for pais in data_json]
continents   =[pais['continents'] for pais in data_json]
area         =[pais['area'] for pais in data_json]
latitute     =[pais['latlng'] for pais in data_json]
linguagens   =[pais.get('languages', 'N/A') for pais in data_json]
moeda        =[pais.get('currencies') for pais in data_json]


pop_paises = pd.DataFrame({
    'Nome' : nome,
    'Nome Oficial': nome_oficial,
    'População' : population,
    'Continente' : continents
})

reg_paises = pd.DataFrame({
    'Nome' : nome,
    'Região' : region,
    'Idioma' : linguagens,
    'Moeda'  : moeda
})

con_paises = pd.DataFrame({
    'Nome' : nome,
    'Continente' : continents,
    'Area' : area,
    'Latitude' : latitute 

})

```
